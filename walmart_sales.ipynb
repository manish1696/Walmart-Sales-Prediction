```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM, SimpleRNN, Dropout, Flatten
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing import sequence
```

```python
df = pd.read_csv('Walmart.csv')
```

```python
df.head()
```

```python
df.Date=pd.to_datetime(df.Date)

df['weekday'] = df.Date.dt.weekday
df['month'] = df.Date.dt.month
df['year'] = df.Date.dt.year
```

```python
df.head()
```

```python
df.drop(['Date'], axis=1, inplace=True)#,'month'

df.head()
```

```python
x=df.drop('Weekly_Sales',axis=1)
y=df['Weekly_Sales']
```

```python
y=np.array(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc =MinMaxScaler()
x = sc.fit_transform(x)
y = sc.fit_transform(y.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model = Sequential()

model.add(LSTM(100, return_sequences= True , input_shape = (9,1)))
model.add(Dropout(0.2))

model.add(LSTM(200, return_sequences= True ))
model.add(Dropout(0.2))

model.add(LSTM(200, return_sequences= True ))
model.add(Dropout(0.2))

model.add(LSTM(100 ))
model.add(Flatten())
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mae', optimizer='adam')
```

```python
model_history = model.fit(x_train, y_train, epochs=100, batch_size=32)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
